```python
#Importing libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Imported Data

diab=pd.read_csv('diabetes.csv')
```

```python
#Verifying wether data has been loaded
#Checking for any impossible zero values in data 

diab.head(10)
```

```python
#Calculating then number of zero values in coloums  

print((diab[['Glucose','BloodPressure','SkinThickness','Insulin','BMI','DiabetesPedigreeFunction','Age']]\
       == 0).sum())
```

```python
#Replacing the zero values with Nan values by using 'np.nan' function

diab[['Glucose','BloodPressure','SkinThickness','Insulin','BMI','Pregnancies']]\
= diab[['Glucose','BloodPressure','SkinThickness','Insulin','BMI','Pregnancies']].replace(0, np.NaN)
```

```python
diab.head(20)
```

```python
#checking whether any of the columns has any value other than numeric value

diab[~diab.applymap(np.isreal).all(1)]
```

```python
#Analyzing the distribution of attributes in the data

diab.describe().transpose()
```

```python
#Analyzing the target coloum 'Outcome' to undertstand the distribution of data among various values

diab.groupby(["Outcome"]).count()
```

```python
diab
```

```python
#Replacing the 'Nan' values in each coloum with the mean of the respective coloums. 

diab['Glucose'].fillna(diab['Glucose'].mean(), inplace=True)
diab['BloodPressure'].fillna(diab['BloodPressure'].mean(), inplace=True)
diab['SkinThickness'].fillna(diab['SkinThickness'].mean(), inplace=True)
diab['Insulin'].fillna(diab['Insulin'].mean(), inplace=True)
diab['BMI'].fillna(diab['BMI'].mean(), inplace=True)    
```
